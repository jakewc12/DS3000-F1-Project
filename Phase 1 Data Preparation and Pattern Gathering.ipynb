```python
#Write your python code to answer question 1. You can add new cells below as needed.

# Run latest version of Python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


circuits = pd.read_csv('circuits.csv')
races = pd.read_csv('races.csv')
drivers = pd.read_csv('drivers.csv')
standings = pd.read_csv('driver_standings.csv')
constructor_results = pd.read_csv('constructor_results.csv')
constructors = pd.read_csv('constructors.csv')
results = pd.read_csv('results.csv')
```

```python
# Turn the nationality of a driver into a country. This allows us to compare a drivers home country with the country of the race
update_countries  = pd.read_csv('demonyms.csv')
for index, row in drivers.iterrows():
    drivers.at[index, 'nationality'] = ''.join(update_countries[update_countries['nationality']==row['nationality']]['country'].values)
```

## Part 5:
Gathering patterns



```python
# Check if drivers driving in their own country perfrom better?
updated = drivers.merge(results)

#drivers with the most points?
winning_drivers = updated.groupby(['driverRef'])['points'].sum()

#get the top 20 drivers
winning_drivers = winning_drivers.sort_values(ascending=False)
winning_drivers = winning_drivers.head(20)

plt.figure(figsize = (15,6))
sns.barplot(x=winning_drivers.values, y=winning_drivers.index, palette="viridis")
plt.title('Drivers With The Most Points on the Grid')
```

As we can see here, Hamilton, Bottas, and Vettel were some drivers that led the group in points.

```python
#Total amount of drivers for each nationality 
df_drivers = drivers.groupby(['nationality'])['driverRef'].count()
df_drivers = df_drivers.rename({'driverRef':'total_drivers'})
df_drivers = df_drivers.sort_values(ascending=False)

df_drivers


x = df_drivers.values
y = df_drivers.index
plt.figure(figsize=(15,6))
plt.title('Nationality of Drivers')
sns.barplot(x=x,y=y,palette='viridis')
plt.xlabel('Count')
```

As we can see here, most of the drivers are of European or American descent. The most common countries were the UK, US, Italy, France, and Germany to name a few. Lesser common countries included China, Indonesia, Poland, and Thailand.

```python
#Which team has the most points since 1950?
constructors
team = constructors.merge(results,on='constructorId')

team_points = team[['name','points','raceId']]

team_points = team_points.groupby(['name'])['raceId'].nunique().sort_values(ascending=False)
#team_points = team_points[team_points['races'] >= 100
team_points = team_points.head(20)

sns.barplot(x=team_points.values, y=team_points.index, palette="viridis")
plt.xlabel('Points')
plt.title('Teams With The Most Points Since 1950')
```

## Looking at any patterns among years and different constructors

```python
constructors_merged = constructor_results.merge(constructors)
constructors_merged = constructors_merged.join(races.set_index('raceId')['year'], on='raceId')

for year in range(1999,2010):
    new = constructors_merged[constructors_merged['year']==year]
    constructors_merged_grouped = new.groupby(['name'])
    constructors_this_results = constructors_merged_grouped['points'].sum().nlargest(5)
    constructors_this_results.plot(kind='bar')
    plt.xlabel('Constructors')
    plt.ylabel('Cumlative points for year '+str(year))
    plt.show()
```
